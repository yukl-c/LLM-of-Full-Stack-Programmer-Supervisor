## Extract the programming language type from a markdown code block indicator 
To display the code in the output with a user-friendly UI, the programming language type is extracted from a markdown code block indicator.

````python
def extract_programming_type(text):
    """
    Extract the programming language type from a markdown code block indicator.
    Example: ```cpp\n -> cpp
    """
    if "```" in text:
        # Split by ``` and take the part after it, then split by newline and take the first part
        try:
            lang_type = text.split("```")[1].split("\n")[0]
            return lang_type
        except IndexError:
            return None
    return None
````

## Model for code generator and guidance system

The maximum amount of tokens in the output for a longer context window and a detailed explanation with codes.
Verbal explanation provided only.

````python
def LLMresponse2(few_shot_prompt, user_ques):
    response = client.models.generate_content(
        model='gemini-2.0-flash',
        config=types.GenerateContentConfig(
            temperature=0.1,
            top_p=0.9,
            max_output_tokens=5000,
            tools=[types.Tool(code_execution=types.ToolCodeExecution())]
        ),
        contents=[few_shot_prompt, user_ques])
    
    # Initialize lang_type to a default value
    lang_type = None
    
    for part in response.candidates[0].content.parts:
        if part.text:
            # Check if this is a code block and extract language type
            if part.text.startswith("```"):
                lang_type = extract_programming_type(part.text)
                print(f"Programming language detected: {lang_type}")
            
            display(Markdown(part.text))
        elif part.executable_code:
            code_text = part.executable_code.code
            # Use the language from executable_code if lang_type isn't set
            if lang_type is None:
                lang_info = part.executable_code.language
            else:
                lang_info = lang_type
                
            print(f"Executable code language: {lang_info}")
            
            display(Markdown(f'```{lang_info}\n{code_text}\n```'))
        elif part.code_execution_result:
            if part.code_execution_result.outcome != 'OUTCOME_OK':
                display(Markdown(f'## Status {part.code_execution_result.outcome}'))
            
            display(Markdown(f'```\n{part.code_execution_result.output}\n```'))
````

## Convert the data type of the user's query to f-string for the code evaluation system
The LLM model used to evaluate the code fetched from the URL with the f-string-typed query, so it is essential to convert the data type to fit the requirement.

````python
# function to convert the user's query from string to fstring
import re

def fstring_query(query):
    list_query = query.split(' ')
    url_regex = r'https?:\/\/[^\s\(\),]+'
    for i in range (len(list_query)):
        if re.match(url_regex, list_query[i]):
            curl_cmd = f"""!curl {list_query[i]}"""
            list_query[i] = curl_cmd
        return f"""{' '.join(list_query)}"""
````

## Model for code evaluation system

````python
def LLMresponse3(few_shot_prompt, user_ques):
    fstr_user_ques = fstring_query(user_ques)
    response=client.models.generate_content(
    model='gemini-2.0-flash',
    contents=[few_shot_prompt, fstr_user_ques])
    # print(response)
    display(Markdown(response.text))
````

## Model for general knowledge system with Grounding

Grounding connects the LLM to reliable external sources, dramatically reducing hallucinations and ensuring responses contain up-to-date, verified information. It can mitigate “AI Hallucination” from Google Search.
I set the response be simple and brief with less than 100 words to save tokens.

````python
def ground_LLMresponse(user_ques):
    print("grounding shown")
    grounding_prompt = "Explain in brief with less than 100 words"
    config_with_search = types.GenerateContentConfig(
            # max_output_tokens=250,
            tools=[types.Tool(google_search=types.GoogleSearch())],
        )
    
    response = client.models.generate_content(
            model='gemini-2.0-flash',
            contents=[grounding_prompt, user_ques],
            config=config_with_search,
        )
        
    rc = response.candidates[0]
        
    display(Markdown(rc.content.parts[0].text))
    chunks = rc.grounding_metadata.grounding_chunks
    if chunks:  # Check if chunks exist
        for chunk in chunks:
            print(f'{chunk.web.title}: {chunk.web.uri}')
````

## Classifing suitable LLM according to the user's query with use of LLM

````python
def llm_classifier(user_query):
    classify_prompt = """
            You are a specialized assistant that helps classify user queries into the most appropriate system for response. 
            Based on the user's query, identify the SINGLE most suitable system from the following options:
            
            1. CODE EVALUATION SYSTEM: For queries asking for code review, assessment, or feedback without URLs. Example: "Is my JavaScript function efficient?" or "Review this algorithm"
            
            2. URL CODE EVALUATION SYSTEM: For queries containing URLs that need code review or assessment. Example: "Review my GitHub repo code" or "Evaluate this code snippet from this link"
            
            3. PRACTICE QUESTION GENERATOR: For queries requesting practice problems, exercises, questions ,or quiz questions. Example: "Give me SQL practice problems" or "JavaScript array exercises"
            
            4. CODE GENERATOR SYSTEM: For queries asking for code creation, examples, or implementation guidance. Example: "Write a Python function to sort a list" or "Show me how to implement a binary tree"
            
            5. CONCEPT GUIDANCE SYSTEM: For general programming concept explanations without specific code requests. Example: "Explain inheritance in OOP" or "What is a closure?"
            
            6. OTHER: For queries unrelated to programming, coding, or IT concepts. Example: "What's the weather like today?" or "Tell me about ancient Roman history"
            
            Analyze the query carefully, focusing on the primary intent rather than secondary elements. 
            If multiple systems could apply, choose the one that best addresses the core request. Respond with ONLY the system name, with no additional explanation.
            
            """
    response = client.models.generate_content(
    model="gemini-2.0-flash",
    config=types.GenerateContentConfig(
            temperature=0.01,),
    contents=[classify_prompt, user_query])

    return response.text
````

## Function Calling to activate appropriate LLM according to the results from the classifier 

````python
def call_appropriate_llm(user_query):
    """
    Function calling wrapper that determines which LLM function to use based on the user's query.
    
    Args:
        few_shot_prompt: The prompt with examples to guide the model
        user_query: The user's question or input
    
    Returns:
        The appropriate LLM response
    """
    try:
        # Get classification from the classifier model
        print("🔍 Classifying query...")
        classify_response = llm_classifier(user_query)
        
        # Clean up response - strip whitespace and convert to uppercase for consistency
        classify_response = classify_response.strip().upper()
        print(f"📊 Classification result: {classify_response}")
        
        # Decision logic based on classification
        if "URL CODE EVALUATION" in classify_response:
            print("🌐 Using code evaluation system for URL...")
            return LLMresponse3(few_shot_prompt4, user_query)
        
        elif "PRACTICE QUESTION" in classify_response:
            print("📝 Using practice question generator...")
            return LLMresponse(few_shot_prompt1, user_query)
        
        elif "CODE GENERATOR" in classify_response:
            print("💻 Using code generator and guidance system...")
            return LLMresponse2(few_shot_prompt3, user_query)
        
        elif "CODE EVALUATION" in classify_response:
            print("🔍 Using code evaluation system...")
            return LLMresponse3(few_shot_prompt4, user_query)
        
        elif "CONCEPT GUIDANCE" in classify_response:
            print("🧠 Using concept guidance system...") 
            return LLMresponse(few_shot_prompt2, user_query)
        
        elif "OTHER" in classify_response:
            print("🌍 Using general knowledge system...") 
            return ground_LLMresponse(user_query)
        
        else:
            # Fallback for unexpected classification results
            print("⚠️ Unrecognized classification. Defaulting to general knowledge system...")
            return ground_LLMresponse(user_query)
            
    except Exception as e:
        # Error handling
        error_message = f"❌ Error in call_appropriate_llm: {str(e)}"
        print(error_message)
        
        # Fallback to the safest option in case of errors
        print("⚠️ Error occurred. Defaulting to general knowledge system...")
        try:
            return ground_LLMresponse(user_query)
        except Exception as inner_e:
            from IPython.display import Markdown, display
            error_text = f"**Error processing your request**: {str(e)}\nSecondary error: {str(inner_e)}"
            display(Markdown(error_text))
            return None
````

## Create a simple interactive chat 
Type 'exit', 'q' or 'quit' to end the conversation.

````python
import time

def simple_interactive_chat():
    print("=== Simple One-Sentence Chat ===")
    print("Type 'exit', 'q' or 'quit' to end")
    print("--------------------------------")
    
    while True:
        # Get user input - one sentence
        user_input = input("\nYou: ")
        
        # Check for exit
        if user_input.lower() in ['exit', 'quit', 'q']:
            print("Conversation ended.")
            break
        start_time = time.time()
        
        # Print the response
        call_appropriate_llm(user_input)
        print("--- %s seconds ---" % (time.time() - start_time))


if __name__ == "__main__":
    simple_interactive_chat()
````
